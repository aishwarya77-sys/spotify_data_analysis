```python
print("Columns in df_tracks:", df_tracks.columns)

# If 'release_date' is found in the columns, proceed with parsing and setting the index
if 'release_date' in df_tracks.columns:
    # Step 1: Handle varying date formats and convert 'release_date' to datetime
    df_tracks['release_date'] = pd.to_datetime(df_tracks['release_date'], errors='coerce', format='%Y-%m-%d')
    df_tracks['release_date'] = df_tracks['release_date'].fillna(pd.to_datetime(df_tracks['release_date'], errors='coerce', format='%Y-%m'))
    df_tracks['release_date'] = df_tracks['release_date'].fillna(pd.to_datetime(df_tracks['release_date'], errors='coerce', format='%Y'))

    # Step 2: Set 'release_date' as the index
    df_tracks.set_index('release_date', inplace=True)

    # Display the head of the DataFrame to verify
    df_tracks.head()
else:
    print("The column 'release_date' is not present in the DataFrame.")
```

```python
df_tracks["artists"].iloc[18]
```

```python
df_tracks['duration']=df_tracks['duration_ms'].apply(lambda x:round(x/1000))
df_tracks.drop('duration_ms',inplace=True,axis=1)
```

```python
df_tracks.duration.head()
```

```python
df_tracks_numeric = df_tracks.apply(pd.to_numeric, errors='coerce')

# Calculate correlation matrix
corr_df = df_tracks_numeric.drop(["key", "mode", "explicit"], axis=1).corr(method="pearson")

# Plotting the heatmap
plt.figure(figsize=(14, 6))
heatmap = sns.heatmap(corr_df, annot=True, fmt=".1g", vmin=-1, vmax=1, center=0, cmap="inferno", linewidths=1, linecolor='black')
heatmap.set_title("Correlation HeatMap Between Variables")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90)
plt.show()
```

```python
df_tracks['dates']=df_tracks.index.get_level_values('release_date')
df_tracks.dates=pd.to_datetime(df_tracks.dates)
years=df_tracks.dates.dt.year
```

```python
sns.displot(years,discrete=True,aspect=2,height=5,kind="hist").set(title="Number of songs per year")
```
